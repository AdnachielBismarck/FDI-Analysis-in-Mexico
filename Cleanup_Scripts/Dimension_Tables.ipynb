### CARGA DE LIBRERÍAS NECESARIAS PARA LA LIMPIEZA

```python
import pandas as pd
import os
```

```python
Ruta_Archivo_Salida = '../Prepared_Data/Joins'
Ruta_Archivo_Entrada = '../Prepared_Data/General/Origin_of_Investment.csv'
```

### TABLA DE ESTADOS - MÉXICO

```python
# Se crea de manera manual una lista con los estados de México
estados = [ "Total general","Aguascalientes", "Baja California", "Baja California Sur", "Campeche", "Chiapas", "Chihuahua", "Ciudad de México", "Coahuila de Zaragoza", "Colima", "Durango", "Estado de México", "Guanajuato", "Guerrero","Hidalgo", "Jalisco", "Michoacán de Ocampo", "Morelos", "Nayarit", "Nuevo León", "Oaxaca", "Puebla", "Querétaro", "Quintana Roo", "San Luis Potosí", "Sinaloa", "Sonora", "Tabasco", "Tamaulipas", "Tlaxcala", "Veracruz de Ignacio de la Llave", "Yucatán", "Zacatecas" ]
# Se convierte a dataframe la lista de estados
Estado = pd.DataFrame({'Estado': estados})
```

```python
# Se procede con su carga a la carpeta seleccionada
Estados = os.path.join(Ruta_Archivo_Salida,'Estado.csv')
Estado.to_csv(Estados,index=False,encoding='utf-8-sig')
```

### TABLA DE AÑOS

```python
# Se crea una lista con los años que tenemos en los datos y se convierte a dataframe
Años = pd.DataFrame({'Año':(list(range(2006,2026)))}) 
```

### TABLA DE PAÍSES DE ORIGEN DEL IED

```python
# Se extraen los países inversores del archivo general creado para los países inversores
Countries = pd.read_csv(Ruta_Archivo_Entrada,encoding='utf-8')
#display(Countries)
Países = Countries['Origen'].unique()
#display(Países)
País = pd.DataFrame(Países)
#display(País)
```
